```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has some builtin datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for plots
sns.set_theme()
```

```python
#visualise the data
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
#load the iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

Scatter plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
#loading the titanic dataset
titanic=sns.load_dataset('titanic')
titanic.head()
```

Count PLot

```python
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

Bar chart

```python
sns.barplot(x='sex',y='survived',hue='class', data=titanic)
```

```python
#house price dataset
import sklearn.datasets
house=sklearn.datasets.fetch_california_housing()

housing=pd.DataFrame(house.data,columns=house.feature_names)
housing['price']=house.target
housing.head()
```

Distribution Plot

```python
sns.distplot(housing['price'])
```

Correlation Matrix

1.Positive correlation

2.negative correaltion

Heat map

```python
correlation=housing.corr()
```

```python
#constructing a heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
