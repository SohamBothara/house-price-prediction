```python
import matplotlib.pyplot as plt
```

```python
import numpy as np  # import numpy to get data for our plots
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# sine wave
plt.plot(x,y)
plt.show
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
```

```python
#cosine wave
plt.plot(x,z)
plt.show
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('cosine wave')
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
```

```python
plt.plot(x,y,'r+')
plt.show
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
```

```python
plt.plot(x,y,'rx')
plt.show
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number of People')
plt.show()
```

Pie Chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```
